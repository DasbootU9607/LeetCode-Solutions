Solution 1 Recursive Function

```python
class Solution:
    def fib(self, n: int) -> int:
        if n==0:
            return 0
        elif n==1:
            return 1
        return self.fib(n-1)+self.fib(n-2)
```

Solution 2 DP with memo array recording

```python
class Solution:
    def fib(self, n: int) -> int:
        if n==0:
            return 0
        memo=[0,1]
        for i in range(1,n):
            memo.append(memo[-1]+memo[-2])
        return memo[-1]
```
