```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        if not headA or not headB:
            return None
        currA=headA
        currB=headB
        lenA=0
        lenB=0
        while currA:
            lenA+=1
            currA=currA.next
        while currB:
            lenB+=1
            currB=currB.next
        currA=headA
        currB=headB
        if lenA>lenB:
            for _ in range(lenA-lenB):
                currA=currA.next
        else:
            for _ in range(lenB-lenA):
                currB=currB.next
        while currA!=currB:
            currA=currA.next
            currB=currB.next
        return currA

```
