
## Hint:

### >Fast and Slow Pointer

### >if Fast meets Slow

### >then exist cycle

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return False
        else:
            slow=head
            fast=head
            while fast and fast.next:
                slow=slow.next
                fast=fast.next.next
                if slow==fast:
                    return True
            return False
```
