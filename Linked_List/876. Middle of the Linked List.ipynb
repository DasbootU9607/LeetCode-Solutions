```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        count=0
        cur=head
        while cur:
            cur=cur.next
            count+=1
        mid=count//2
        point=head
        for _ in range(mid):
            point=point.next
        return point
```
