### Solution 1: Using a List

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if not head or head.next:
            return True
        list=[]
        cur=head
        while cur:
            list.append(cur.val)
            cur=cur.next

        mid=len(list)//2
        left=list[:mid]
        right=list[::-1][:mid]
        return left==right
```

### Solution 2: Using fast and slow pointer


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return True
        fast=slow=head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next

        prev=None
        cur=slow
        while cur:
            next_node=cur.next
            cur.next=prev
            prev=cur
            cur=next_node
        left_ptr=head
        right_ptr=prev
        while right_ptr:
            if left_ptr.val!=right_ptr.val:
                return False
            left_ptr=left_ptr.next
            right_ptr=right_ptr.next
        return True
```
