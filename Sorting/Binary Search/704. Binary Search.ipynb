```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        head,tail=0,len(nums)-1

        while head<=tail:
            mid=(head+tail)//2
            if nums[mid]==target:
                return mid
            elif nums[mid]<target:
                head=mid+1

            elif nums[mid]>target:
                tail=mid-1
        return -1
```
