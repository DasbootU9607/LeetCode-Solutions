```python
class Trie(object):

    def __init__(self):
        self.child = {}
        self.isEnd = False

    def insert(self, word):
        node = self
        for token in word:
            if token not in node.child:
                node.child[token] = Trie()
            node = node.child[token]
        node.isEnd = True

    def search(self, word):
        node = self
        for token in word:
            if token not in node.child:
                return False
            node = node.child[token]
        return node.isEnd

    def startsWith(self, prefix):
        node = self
        for token in prefix:
            if token not in node.child:
                return False
            node = node.child[token]
        return True
```
