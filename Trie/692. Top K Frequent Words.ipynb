```python
class Trie:
    def __init__(self):
        self.child={}
        self.word=None
    def insert(self,word):
        node=self
        for char in word:
            if char not in node.child:
                node.child[char]=Trie()
            node=node.child[char]
        node.word=word
class Solution(object):
    def topKFrequent(self, words, k):
        """
        :type words: List[str]
        :type k: int
        :rtype: List[str]
        """
        counts={}
        for w in words:
            counts[w]=counts.get(w,0)+1

        root=Trie()
        for w in counts:
            root.insert(w)
        sort_word=[]
        def dfs(node):
            if node.word:
                sort_word.append(node.word)
            for char in sorted(node.child.keys()):
                dfs(node.child[char])
        dfs(root)
        sort_word.sort(key=lambda x:counts[x],reverse=True)
        return sort_word[:k]
```
