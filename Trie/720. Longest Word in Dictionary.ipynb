```python
class TrieNode:
    def __init__(self):
        self.child={}
        self.isEnd=False

class Solution(object):
    def longestWord(self, words):
        """
        :type words: List[str]
        :rtype: str
        """
        root=TrieNode()
        for word in words:
            node=root
            for char in word:
                if char not in node.child:
                    node.child[char]=TrieNode()
                node=node.child[char]
            node.isEnd=True

        res=""
        for word in words:
            node=root
            valid=True
            for char in word:
                node=node.child[char]
                if not node.isEnd:
                    valid=False
                    break
            if valid:
                if len(word)>len(res) or (len(word)==len(res) and word<res):
                    res=word
        return res
```
