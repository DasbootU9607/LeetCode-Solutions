### Use Trie type to store the root and wirt e out the finding root function for each word. Then use split function to seperate each word in sentence.

```python
class Trie:
    def __init__(self):
        self.child={}
        self.isEnd=False
    def insert(self,word):
        node=self
        for char in word:
            if char not in node.child:
                node.child[char]=Trie()
            node=node.child[char]
        node.isEnd=True
    def find_root(self,word):
        node=self
        prefix=""
        for char in word:
            if char not in node.child:
                break
            node=node.child[char]
            prefix+=char
            if node.isEnd:
                return prefix
        return word

class Solution(object):
    def replaceWords(self, dictionary, sentence):
        """
        :type dictionary: List[str]
        :type sentence: str
        :rtype: str
        """
        trie=Trie()
        for root in dictionary:
            trie.insert(root)
        words=sentence.split(" ")
        res=[]
        for word in words:
            res.append(trie.find_root(word))
        return " ".join(res)
```
