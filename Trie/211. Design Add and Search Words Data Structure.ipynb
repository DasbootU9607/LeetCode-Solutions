### Using recursive function dfs as helper to find at least one solution that meets the requirements

```python
class WordDictionary(object):

    def __init__(self):
        self.child={}
        self.isEnd=False

    def addWord(self, word):
        """
        :type word: str
        :rtype: None
        """
        node=self
        for char in word:
            if char not in node.child:
                node.child[char]=WordDictionary()
            node=node.child[char]
        node.isEnd=True

    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """
        return self._dfs(word,self)

    def _dfs(self,word,node):
        for i,char in enumerate(word):
            if char==".":
                for child in node.child.values():
                    if self._dfs(word[i+1:],child):
                        return True
                return False
            else:
                if char not in node.child:
                    return False
                node=node.child[char]
        return node.isEnd



# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```
