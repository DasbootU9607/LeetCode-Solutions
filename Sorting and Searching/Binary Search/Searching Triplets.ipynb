### Problem 2: Arithmetic Triplets in a Sorted Array

**Description:**
Given an integer $N$ and an ascending sorted array $A$ consisting of $N$ integers, your task is to implement the function `solve(N, A)` to find the number of triplets $(j, i, k)$ that satisfy an arithmetic progression relationship.

Specifically, you need to count the number of index combinations where $0 \le j < i < k < N$ and the elements satisfy the following condition:
$$A[j] + A[k] = 2 \times A[i]$$

**Input Format:**
- The input is provided as a sequence of numbers. The first number is $N$ (the size of the array).
- The subsequent $N$ numbers represent the elements of the sorted array $A$.

**Output Format:**
- Return and print a single integer representing the total number of valid triplets.

```python
def solve(N, A):
#add you codes
    count=0
    for i in range(1,N-1):
        left,right=0,N-1

        while left<i<right:

            if 2*A[i]==A[left]+A[right]:
                count+=1
                left+=1
                right-=1
            elif 2*A[i]>A[left]+A[right]:
                left+=1
            else:
                right-=1
    return count
```

```python
# main function
import sys
input = sys.stdin.read
data = input().split()

N = int(data[0])
A = list(map(int, data[1:]))
print(solve(N,A))
```
