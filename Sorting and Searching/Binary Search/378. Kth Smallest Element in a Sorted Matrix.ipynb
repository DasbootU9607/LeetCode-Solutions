```python
class Solution(object):
    def kthSmallest(self, matrix, k):
        """
        :type matrix: List[List[int]]
        :type k: int
        :rtype: int
        """
        n=len(matrix)
        low,high=matrix[0][0],matrix[n-1][n-1]

        def count(mid):
            col,row=0,n-1
            count=0
            while col<n and row>=0:
                if mid>=matrix[row][col]:
                    col+=1
                    count+=(row+1)
                else:
                    row-=1
            return count

        while low<high:
            mid=(low+high)//2
            if count(mid)<k:
                low=mid+1
            else:
                high=mid
        return low
```
