```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left,right=0,len(nums)-1

        rotate=0
        while left<=right:
            mid=(left+right)//2

            if nums[mid]==target:
                return mid

# first find the sorted part of the list, then apply binary search
            if nums[left]<=nums[mid]:   # the left part of the list is sorted
                if nums[left]<=target<nums[mid]:
                    right=mid-1
                else:
                    left=mid+1
            else:                      # the right part of the list is sorted
                if nums[mid]<target<=nums[right]:
                    left=mid+1
                else:
                    right=mid-1
        return -1
```
