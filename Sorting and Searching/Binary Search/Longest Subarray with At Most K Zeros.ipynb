### Problem 1 Longest Subarray with At Most K Zeros

**Description:**

You are given an unsorted array containing only 0s and 1s, and an integer K. Write a function to determine the length of the longest contiguous subarray that contains at most K zeros.

Note: A subarray must be contiguous. For example, for a given array

[1 0 1 1 0], [0 1 1] is a subarray, but [1 1 1 0] is not.

Function Prototype:

Python

def longest_subarray(arr, K):

Example Input:

1 0 1 1 0 1 1 0

1

Example Output:

5

```python
def longest_subarray(arr, K):
#add you codes
    left=0
    zero_count=0
    max_len=0
    for right in range(len(arr)):
        if arr[right]==0:
            zero_count+=1
        while zero_count>K:
            if arr[left]==0:
                zero_count-=1
            left+=1
        curr_len=right-left+1
        max_len=max(max_len,curr_len)
    return max_len
```

```python
arr = list(map(int, input().split()))
K = int(input())

print(longest_subarray(arr, K))
```
