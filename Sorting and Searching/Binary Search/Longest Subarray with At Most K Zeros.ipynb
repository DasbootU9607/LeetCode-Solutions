````python
def longest_subarray(arr, K):
#add you codes
    left=0
    zero_count=0
    max_len=0
    for right in range(len(arr)):
        if arr[right]==0:
            zero_count+=1
        while zero_count>K:
            if arr[left]==0:
                zero_count-=1
            left+=1
        curr_len=right-left+1
        max_len=max(max_len,curr_len)
    return max_len
````

````python
arr = list(map(int, input().split()))
K = int(input())

print(longest_subarray(arr, K))
````
