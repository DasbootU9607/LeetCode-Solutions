### Use set to represent hash table, search in set is O(1)

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set=set(nums)
        length=0
        for num in num_set:
            if num-1 not in num_set:   # only enter when the "not" is met, thus counting from the smallest number of the sequence
                curr_num=num
                curr_length=1
                while curr_num+1 in num_set:
                    curr_num+=1
                    curr_length+=1
                length=max(length,curr_length)
        return length
```
