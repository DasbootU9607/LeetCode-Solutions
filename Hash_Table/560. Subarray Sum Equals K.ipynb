### Prefix sum num=[2,5,4,-2]

### prefix_sum=[2,7,11,9]  =>sum from in dex i to j is prefix_sum[j]-prefix_sum[i]

### use dictionary (hash table) to record prefix_sum

#### prefix_sum[j]-prefix_sum[i]=k   =>  prefix_sum[j]-k=prefix_sum[i]

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        freq={0:1}  # prefix_sum 0 has appeared once
        pre_sum=0
        count=0 # total number of intervals s.t. sum of interval=k
        for num in nums:
            pre_sum+=num

            if pre_sum-k in freq:
                count+=freq[pre_sum-k]
            freq[pre_sum]=freq.get(pre_sum,0)+1 # in dictionary: get(key,default) if no key exists,then fetch 0, if key exits, then fetch key
        return count
```
