```python
class MyHashMap(object):

    def __init__(self):
        self.base=7
        self.table=[[] for _ in range(self.base)]
    def func(self,key):
        return key%self.base
    def put(self, key, value):
        """
        :type key: int
        :type value: int
        :rtype: None
        """
        index=self.func(key)
        for item in self.table[index]:
            if item[0]==key:
                item[1]=value
                return
        self.table[index].append([key,value])


    def get(self, key):
        """
        :type key: int
        :rtype: int
        """
        index=self.func(key)
        for item in self.table[index]:
            if item[0]==key:
                return item[1]
        return -1


    def remove(self, key):
        """
        :type key: int
        :rtype: None
        """
        index=self.func(key)
        for i, item in enumerate(self.table[index]):
            if item[0]==key:
                self.table[index].pop(i)
                return



# Your MyHashMap object will be instantiated and called as such:
# obj = MyHashMap()
# obj.put(key,value)
# param_2 = obj.get(key)
# obj.remove(key)
```
