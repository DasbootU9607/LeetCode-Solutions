```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        prev_map={}
        for i,num in enumerate(nums):  #enumerate: get index and data at the same time
            comp=target-num
            if comp in prev_map:
                return [prev_map[comp],i]
            prev_map[num]=i
        return []
```
