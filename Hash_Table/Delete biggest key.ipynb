### Problem 3: Hash Table with Closed Addressing (Chaining)

**Description:**
You are tasked with completing the implementation of a Hash Table that resolves collisions using **closed addressing (separate chaining)**. You are provided with the basic `Node` and `HashTable` classes, as well as a hash function `index = key % hSize`.

You need to implement the following two functions:

1. `hash_insert(ht, key)`:
   - Inserts a given `key` into the hash table `ht`.
   - If the key already exists in the table (i.e., it is a duplicate), do **not** insert it.
   - When resolving collisions, new nodes should be inserted at the beginning of the corresponding linked list (Head Insertion).
   - Returns `1` if the insertion is successful, and `0` if the key is a duplicate or the table size is 0.

2. `hash_delete_biggest(ht)`:
   - Traverses the entire hash table to find the maximum key currently stored.
   - Deletes the node containing this maximum key from its respective linked list.
   - Returns `1` if the deletion is successful, and `0` if the hash table is empty or the maximum key cannot be found.

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.next = None

class HashTable:
    def __init__(self, hSize=0):
        self.hSize = hSize #the number of hash slots
        self.nSize = 0 #the number of keys
        self.table = [None] * self.hSize

def hash_func(key, hSize):
    return key % hSize

def hash_search(ht, key):
    if ht.hSize == 0:
        return None
    index = hash_func(key, ht.hSize)
    curr = ht.table[index]
    while curr:
        if curr.key == key:
            return curr
        curr = curr.next
    return None

def hash_print(ht):
    for i in range(ht.hSize):
        print(f"{i}:", end=" ")
        curr = ht.table[i]
        while curr:
            print(f"{curr.key} ->", end=" ")
            curr = curr.next
        print()
```

```python
def hash_insert(ht, key):
    if hash_search(ht,key):
        return 0
    if ht.hSize==0:
        return 0
    index=hash_func(key,ht.hSize)
    new_node=Node(key)
    new_node.next=ht.table[index]
    ht.table[index]=new_node
    ht.nSize+=1
    return 1


def hash_delete_biggest(ht):
    #add you implementations
    if ht.hSize==0 or ht.nSize==0:
        return 0
    max_num=-1
    max_index=-1
    for i in range(ht.hSize):
        curr=ht.table[i]
        while curr:
            if curr.key>max_num:
                max_num=curr.key
                max_index=i
            curr=curr.next
    if max_num==-1:
        return 0

    curr=ht.table[max_index]
    prev=None
    while curr:
        if curr.key==max_num:
            if prev==None:
                ht.table[max_index]=curr.next
            else:
                prev.next=curr.next
            ht.nSize-=1
            return 1
        prev=curr
        curr=curr.next
    return 0
```

```python
# main function
def main():
    ht = None

    import sys
    input = sys.stdin.read
    data = list(map(int, input().split()))
    i = 0

    print("============= Hash Table ============")
    print("|1. Create a hash table                   |")
    print("|2. Insert a key to the hash table        |")
    print("|3. Search a key in the hash table        |")
    print("|4. Delete biggest key in the hash table  |")
    print("|5. Print the hash table                  |")
    print("|6. Quit                                  |")
    print("=====================================")

    print("Enter Selection: ", end="")
    while i < len(data):
        opt = data[i]
        i += 1

        if opt == 1:
            print("Enter the size of hash table:")
            size = data[i]
            i+=1
            ht = HashTable(size)
            print("HashTable is created.")

        elif opt == 2:
            if ht is None:
                print("HashTable not created yet.")
                continue
            print("Enter a key to be inserted:")
            key = data[i]
            i+=1
            if hash_insert(ht, key):
                print(f"{key} is inserted.")
            else:
                print(f"{key} is a duplicate. No key is inserted.")
        elif opt == 3:
            if ht is None:
                print("HashTable not created yet.")
                continue
            print("Enter a key for searching in the HashTable:")
            key = data[i]
            i+=1
            if hash_search(ht, key):
                print(f"{key} is found.")
            else:
                print(f"{key} is not found.")
        elif opt == 4:
            if ht is None:
                print("HashTable not created yet.")
                continue
            if hash_delete_biggest(ht):
                print("Biggest key is deleted.")
            else:
                print("Biggest key is not existing.")
        elif opt == 5:
            if ht is None:
                print("HashTable not created yet.")
                continue
            hash_print(ht)
        elif opt == 6:
            print("Exiting.")
            break
        else:
            print("Invalid option.")
        print("Enter Selection: ", end="")

if __name__ == "__main__":
    main()
```
