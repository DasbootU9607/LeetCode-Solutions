```python
class MyHashSet(object):

    def __init__(self):
        self.base = 7
        self.hash_table = [[] for _ in range(self.base)]
    def _func(self, key):
        return key % self.base

    def add(self, key):
        """
        :type key: int
        :rtype: None
        """
        index = self._func(key)
        if key not in self.hash_table[index]:
            self.hash_table[index].append(key)

    def remove(self, key):
        """
        :type key: int
        :rtype: None
        """
        index = self._func(key)
        if key in self.hash_table[index]:
            self.hash_table[index].remove(key)

    def contains(self, key):
        """
        :type key: int
        :rtype: bool
        """
        index = self._func(key)
        return key in self.hash_table[index]
```
