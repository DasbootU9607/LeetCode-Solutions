```python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        res=[]
        path=[]
        used=[False]*len(nums)
        def backtracking():
            if len(path)==len(nums):
                res.append(list(path))
                return
            for i in range(len(nums)): # iterate through every element
                if used[i]:
                    continue
                used[i]=True
                path.append(nums[i])
                backtracking()

                path.pop()
                used[i]=False   # restore for next iteration
        backtracking()
        return res
```
