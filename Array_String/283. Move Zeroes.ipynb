```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        scan=0
        for i in range(len(nums)):
            if nums[i]!=0:
                nums[scan],nums[i]=nums[i],nums[scan]
                scan+=1
        return nums
```
