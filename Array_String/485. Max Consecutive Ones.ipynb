```python
class Solution(object):
    def findMaxConsecutiveOnes(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count=0
        sum=0
        max_ans=0
        for num in nums:
            tmp=sum
            sum+=num
            if sum!=tmp:
                count+=1
            else:
                if count>max_ans:
                    max_ans=count
                count=0
        return max(max_ans,count)
```
