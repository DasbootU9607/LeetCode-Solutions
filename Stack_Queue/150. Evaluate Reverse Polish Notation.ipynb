### Calculation of Postfix expression using stack

```python
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        operators={
            "*": lambda a,b: a*b,
            "/": lambda a,b: int(a/b),
            "+": lambda a,b: a+b,
            "-": lambda a,b: a-b}
        stack=[]
        for token in tokens:
            if token in operators:
                num2=stack.pop()
                num1=stack.pop()
                ans=operators[token](num1,num2)
                stack.append(ans)
            else:
                stack.append(int(token))
        result=stack.pop()
        return result
```
