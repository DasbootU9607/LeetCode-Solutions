### Solution 1

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack=[]
        balanced=True
        index=0
        while index<len(s) and balanced:
            symbol=s[index]
            if symbol in "({[":
                stack.append(symbol)
            elif symbol in "]})":
                if stack==[]:
                    balanced=False
                else:
                    top=stack.pop()
                    if not self.matches(top,symbol):
                        balanced=False
            index+=1
        return balanced and stack==[]
    def matches(self,top,symbol):
        opens="({["
        closers=")}]"
        return opens.index(top)==closers.index(symbol)
```

### Solution 2

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack=[]
        mapping={"{":"}","(":")","[":"]"}
        for symbol in s:
            if symbol in mapping:
                stack.append(symbol)
            else:
                if stack==[]:
                    return False
                top=stack.pop()
                if mapping[top]!=symbol:
                    return False
        return stack==[]
```

### Solution 3

```python
class Solution:
    def isValid(self, s: str) -> bool:
        mapping = {')': '(', '}': '{', ']': '['}
        stack = []
        for char in s:
            if char in mapping:
                if not stack or stack.pop() != mapping[char]:
                    return False
            else:
                stack.append(char)
        return not stack
```
