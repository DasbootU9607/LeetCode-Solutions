Solution 1 Using Queue

```python
class Solution:
    def timeRequiredToBuy(self, tickets: List[int], k: int) -> int:
        line=Queue()
        time=0
        for i in range(len(tickets)):
            node=Node(tickets[i],i)
            line.enqueue(node)
        while not line.is_empty():
            person=line.dequeue()
            time+=1
            person.val-=1
            if person.val==0 and person.id==k:
                return time
            if person.val>0:
                person.next=None
                line.enqueue(person)
        return time



class Queue(object):
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
    def enqueue(self,node):
        if self.is_empty():
            self.head=node
            self.tail=node
        else:
            self.tail.next=node
            self.tail=node
        self.size+=1
    def dequeue(self):
        if self.is_empty():
            return None
        tar=self.head
        self.head=self.head.next
        if self.head is None:
            self.tail=None
        tar.next=None
        self.size-=1
        return tar
    def is_empty(self):
        return self.size==0
class Node(object):
    def __init__(self,val,id):
        self.val=val
        self.next=None
        self.id=id



```
