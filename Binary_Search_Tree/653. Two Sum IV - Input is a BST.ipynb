### Solution 1 Using array to inorder traverse the bst and use brute force in array

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def findTarget(self, root: Optional[TreeNode], k: int) -> bool:
        arr=[]
        def helper(root):
            if not root:
                return
            helper(root.left)
            arr.append(root.val)
            helper(root.right)
        helper(root)
        for i in range(len(arr)):
            for val in arr[i+1:len(arr)+1]:
                if arr[i]+val==k:
                    return True
        else:
            return False
```

### Solution 2 Using array to inorder traverse the bst and use two pointer in array

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def findTarget(self, root: Optional[TreeNode], k: int) -> bool:
        arr=[]
        def helper(root):
            if not root:
                return
            helper(root.left)
            arr.append(root.val)
            helper(root.right)
        helper(root)
        left=0
        right=len(arr)-1
        while left<right:
            sum=arr[left]+arr[right]
            if sum==k:
                return True
            elif sum<k:
                left+=1
            else:
                right-=1
        return False
```
